```
import pandas as pd
```

```
orientcoes_df = pd.read_excel(r'/content/drive/MyDrive/Colab Notebooks/Projetos/Bancada empresarial/planilha_consolidada(proposições+orientações).xlsx')
votacoes_df = pd.read_excel(r'/content/drive/MyDrive/Colab Notebooks/Projetos/Bancada empresarial/BD_empresários_votações_1998_2014.xlsx')
display(orientcoes_df, votacoes_df)
```

```
base_votacoes = votacoes_df.drop(columns='PROCV')
base_orientacoes = orientcoes_df

display(base_votacoes.info(null_counts=True), base_orientacoes.info(null_counts=True))
```

```
base_votacoes['Partido do Parlamentar'] = base_votacoes['Partido do Parlamentar'].str.upper()
#base_votacoes = base_votacoes.rename(columns={'Partido do Parlamentar': 'Partido_bancada'})
#base_votacoes = base_votacoes.rename(columns={'Partido_bancada': 'Partido do Parlamentar'})

base_orientacoes['Sigla da bancada'] = base_orientacoes['Sigla da bancada'].str.upper()
#base_orientacoes = base_orientacoes.rename(columns={'Sigla da bancada': 'Partido_bancada'})
```

Depara bancadas

```
base_depara_bancadas = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Projetos/Bancada empresarial/DEPARA_bancadas.xlsx')
base_depara_bancadas['DE'] = base_depara_bancadas['DE'].str.upper()
base_depara_bancadas['PARA'] = base_depara_bancadas['PARA'].str.upper()


for i, item in enumerate(base_depara_bancadas['PARA']):
    tupla = tuple(item.split('/'))
    base_depara_bancadas.loc[i, 'bancada_tuplas'] = tupla

base_orientacoes_depara = base_orientacoes.merge(base_depara_bancadas, how='left', left_on='Sigla da bancada', right_on='DE')
base_orientacoes_depara = base_orientacoes_depara.rename(columns={'Partido_bancada': 'Partido_bancada_original','bancada_tuplas': 'Partido_bancada_tupla'})
display(base_orientacoes_depara.info())
```

##Merge

```
base_votacoes_merge = base_votacoes.merge(base_orientacoes_depara.explode('Partido_bancada_tupla', ignore_index=False), 
                                          how='left', 
                                          left_on=['Partido do Parlamentar', 'Código da Proposição'],
                                          right_on=['Partido_bancada_tupla', 'Código da Proposição'],
                                         )

display(base_votacoes_merge.info(null_counts=True))
```

```
base_votacoes_merge_null = base_votacoes_merge[base_votacoes_merge['Orientação da Bancada'].isna() == True]
base_votacoes_merge_notnull = base_votacoes_merge[base_votacoes_merge['Orientação da Bancada'].isna() != True]

#display(base_votacoes_merge_null, base_votacoes_merge_notnull)
base_votacoes_merge_notnull.info(null_counts = True)

base_merge_teste = base_votacoes_merge_notnull[base_votacoes_merge_notnull['Código da Proposição'] == 127970]
base_merge_teste = base_merge_teste.drop_duplicates()
base_merge_teste.to_excel('/content/drive/MyDrive/Colab Notebooks/Projetos/Bancada empresarial/BDs/base_merge_teste.xlsx')

base_votacoes_teste = base_votacoes[base_votacoes['Código da Proposição'] == 127970]
base_votacoes_teste.to_excel('/content/drive/MyDrive/Colab Notebooks/Projetos/Bancada empresarial/BDs/base_votacoes_teste.xlsx')

base_orientacoes_teste = base_orientacoes[base_orientacoes['Código da Proposição'] == 127970]
base_orientacoes_teste.to_excel('/content/drive/MyDrive/Colab Notebooks/Projetos/Bancada empresarial/BDs/base_orientacoes_teste.xlsx')
```
